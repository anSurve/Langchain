## Indexing: Loading document with pyPDF loader

## Loading PDF file

```python
from langchain_chroma import Chroma
from langchain_community.document_loaders import PyPDFLoader
import copy
```

```python
loader_pdf = PyPDFLoader("data/Introduction_to_Data_and_Data_Science.pdf")
```

```python
pages_pdf = loader_pdf.load()
```

```python
pages_pdf
```

```python
len(pages_pdf)
```

```python
pages_pdf_cut = copy.deepcopy(pages_pdf)
```

```python
' '.join(pages_pdf_cut[0].page_content.split())
```

```python
for i in pages_pdf_cut:
    i.page_content = ' '.join(i.page_content.split())
```
