## Runnable Lambda

```python
from langchain_core.runnables import RunnableLambda, chain
```

```python
find_sum = lambda x: sum(x)
```

```python
find_sum([1, 3, 5])
```

```python
find_square = lambda x: x ** 2
```

```python
find_square(8)
```

```python
runnable_sum = RunnableLambda(lambda x: sum(x))
```

```python
runnable_sum.invoke([1,3,5])
```

```python
runnable_square = RunnableLambda(lambda x: x**2)
```

```python
runnable_square.invoke(8)
```

```python
chain = runnable_sum | runnable_square
```

```python
chain.invoke([1, 3, 5])
```

```python
chain.get_graph().print_ascii()
```

```python
def find_sum(x):
    return sum(x)
```

```python
def find_square(x):
    return x**2
```

```python
chain1 = RunnableLambda(find_sum) | RunnableLambda(find_square)
```

```python
chain1.invoke([1, 2, 5])
```

```python
@chain
def runnable_sum(x):
    return sum(x)

@chain
def runnable_square(x):
    return x**2
```

```python
type(runnable_sum), type(runnable_square)
```

```python
chain2 = runnable_sum | runnable_square
```

```python
chain2.invoke([1, 3, 5])
```
