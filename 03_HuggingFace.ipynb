```python
from transformers import pipeline
```

```python
sentiment_classifier = pipeline('sentiment-analysis')
```

```python
sentiment_classifier('I am so excited te be learning about large language models')
```

```python
ner = pipeline('ner', model='dslim/bert-base-NER')
```

```python
ner('Her name is Anna and she works in New York City for Morgan Stanly')
```

```python
zeroshot_classifier = pipeline('zero-shot-classification', model='facebook/bart-large-mnli')
```

```python
sequence_to_classify = "one day I will see the world"
candidate_labels = ['cooking', 'travel', 'dancing']
zeroshot_classifier(sequence_to_classify, candidate_labels)
```

 ## Pre Trained Tokenizers


```python
from transformers import AutoTokenizer
```

```python
model = "bert-base-uncased"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
sentence = "I am so excited to be learning about large language models"
```

```python
input_ids = tokenizer(sentence)
```

```python
print(input_ids)
```

```python
tokens = tokenizer.tokenize(sentence)
```

```python
print(tokens)
```

```python
token_ids = tokenizer.convert_tokens_to_ids(tokens)
```

```python
print(token_ids)
```

```python
decoded_ids = tokenizer.decode(token_ids)
```

```python
print(decoded_ids)
```

```python
tokenizer.decode(101)
```

```python
tokenizer.decode(102)
```

```python
model2 = "xlnet-base-cased"
```

```python
tokenizer2 = AutoTokenizer.from_pretrained(model2)
```

```python
input_ids2 = tokenizer2(sentence)
```

```python
print(input_ids2)
```

```python
tokens2 = tokenizer2.tokenize(sentence)
print(tokens2)
```

```python
token_ids2 = tokenizer2.convert_tokens_to_ids(tokens2)
print(token_ids2)
```
