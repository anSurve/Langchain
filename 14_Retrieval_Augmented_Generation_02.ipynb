## Document splitting by no of characters

```python
from langchain_community.document_loaders import Docx2txtLoader
from langchain_text_splitters.character import CharacterTextSplitter
```

```python
loader_docx = Docx2txtLoader("data/Introduction_to_Data_and_Data_Science.docx")
pages_docx = loader_docx.load()
for i in range(len(pages_docx)):
    pages_docx[i].page_content = ' '.join(pages_docx[i].page_content.split())
```

```python
len(pages_docx[0].page_content)
```

```python
character_splitter= CharacterTextSplitter(separator = '', chunk_size = 500, chunk_overlap = 0)
```

```python
pages_char_split = character_splitter.split_documents(pages_docx)
```
