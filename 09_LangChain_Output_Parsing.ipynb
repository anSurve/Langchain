```python
from configs import API_KEY, DEFAULT_MODEL
import os
import openai
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.messages import HumanMessage
from langchain_core.output_parsers import CommaSeparatedListOutputParser
# from langchain_core.output_parsers import DateTimeOutputParser
```

```python
os.environ["OPENAI_API_KEY"] = API_KEY
# openai.api_key = API_KEY
openai.api_key = os.getenv('OPENAI_API_KEY')
```

```python
chat= ChatOpenAI(
    model = 'gpt-4',
    model_kwargs = {'seed':365},
    temperature = 0,
    max_tokens = 100
)
```

```python
message_h = HumanMessage(content = f"""I have recentl adopted a dog, can you suggest me a few dog names

{CommaSeparatedListOutputParser().get_format_instructions()}
""")
```

```python
response = chat.invoke([message_h])
```
